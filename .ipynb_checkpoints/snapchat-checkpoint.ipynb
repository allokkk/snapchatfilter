```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```

```python
img=cv2.imread('Jamie_Before.jpg')
rgbimg=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
glasses=cv2.imread('glasses.png',cv2.IMREAD_UNCHANGED)
mustache=cv2.imread('mustache.png',cv2.IMREAD_UNCHANGED)
plt.imshow(rgbimg)
plt.show()
```

```python
eyes_cascade=cv2.CascadeClassifier("frontalEyes35x16.xml")
nose_cascade=cv2.CascadeClassifier("Nose18x15.xml")
```

```python
def detecteye(img_):
    eyes=eyes_cascade.detectMultiScale(img_,1.3,5)
    for eye in eyes:
        x1,y1,w1,h1=eye 
        return eye
def detectnose(img_):
    noses=nose_cascade.detectMultiScale(img_,1.3,5)
    for nose in noses:
        x2,y2,w2,h2=nose
        return nose
```

```python
eye_=detecteye(rgbimg)
nose_=detectnose(rgbimg)
print(eye_)
print(nose_)
```

```python
glasses=cv2.resize(glasses,dsize=(170,371))
mustache=cv2.resize(mustache,dsize=(116,139))
```

```python
glasses_= cv2.cvtColor(glasses,cv2.COLOR_BGR2RGBA)
plt.imshow(glasses_)
plt.show()
```

```python
mustache = cv2.cvtColor(mustache,cv2.COLOR_BGR2RGBA)
plt.imshow(mustache)
plt.show()
```

```python
y1_offset =nose_[1]+50 # 50 to push mustach down
x1_offset= nose_[0]   #312 487 107 129

y1a, y2a = y1_offset, y1_offset + nose_[2]
x1a, x2a = x1_offset, x1_offset + nose_[3]

alpha_sa = mustache[:,:,2] /255
alpha_la = 1.0 - alpha_sa

for ca in range(0, 2):
    rgbimg[y1a:y2a, x1a:x2a, ca] = (alpha_sa * mustache[:, :, ca] +alpha_la * rgbimg[y1a:y2a, x1a:x2a, ca])
```

```python
y_offset =eye_[1]
x_offset= eye_[0]   # [189 336 170 371]

y1, y2 = y_offset, y_offset + eye_[2]
x1, x2 = x_offset, x_offset + eye_[3]

alpha_s = glasses[:, :, 3] /255
alpha_l = 1.0 - alpha_s

for c in range(0, 3):
    rgbimg[y1:y2, x1:x2, c] = (alpha_s * glasses[:, :, c] +
                              alpha_l * rgbimg[y1:y2, x1:x2, c])
```
